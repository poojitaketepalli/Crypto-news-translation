# Coindesk website scraping

```python
from bs4 import BeautifulSoup
import requests
import urllib.request,sys,time
import pandas as pd
import csv
```

```python
url="https://www.coindesk.com/"
itemsfull=[]
p = requests.get(url)
soup=BeautifulSoup(p.content,'html.parser')
content = soup.find_all('div',class_='article-cardstyles__AcTitle-q1x8lc-1 bwXBTf')
heading = ["News"]
for item in content:
    items=[]
    news = item.find('a',class_ = 'headline')
    items.append(news.text)
    itemsfull.append(items)
```

```python
df = pd.DataFrame(itemsfull, columns =['News'])
df.head()
```

```python
df['News'][1]
```

# Translation

```python
from googletrans import Translator
```

```python
translator = Translator()
```

```python
for i in range(len(df)):
    translation = translator.translate(df["News"][i],dest='te',src='en')
    translation_1 = translator.translate(df["News"][i],dest='ta',src='en')
    print(df["News"][i]," : ",translation.text,"\n",translation_1.text)
    
```
